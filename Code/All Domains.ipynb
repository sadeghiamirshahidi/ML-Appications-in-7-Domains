```python
import matplotlib.pyplot as plt
```

```python
# Data
ml_methods = [
    "Healthcare", 
    "Industry 4.0 and Manufacturing", 
    "Additive Manufacturing", 
    "Finance", 
    "Autonomous Systems and Robotic", 
    "Computer Vision", 
    "Environmental & Sustainability"
]
papers = [21, 147, 19, 115, 24, 25, 158]
```

```python

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    papers,
    labels=ml_methods,
    autopct="%1.1f%%",
    startangle=140,
    shadow=True
)
plt.title("Distribution of Machine Learning Papers Across Domains", fontsize=14, fontweight="bold")
plt.show()
```
