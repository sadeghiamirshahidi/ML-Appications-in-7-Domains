```python
# Import Required Packages
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
ml_method_types = ["Supervised", "Unsupervised", "Reinforcement", "Deep Learning"]
numbers = np.array([87, 6, 0, 16])
```

```python
# Sort data in descending order for Pareto
sorted_indices = np.argsort(numbers)[::-1]
ml_method_types_sorted = np.array(ml_method_types)[sorted_indices]
numbers_sorted = numbers[sorted_indices]

# Calculate cumulative percentage
cumulative = np.cumsum(numbers_sorted)
cumulative_percent = 100 * cumulative / cumulative[-1]
```

### Pareto Chart for ML Methods in Healthcare

```python
# Create figure and twin axis
fig, ax1 = plt.subplots(figsize=(8, 6))

# Bar chart (frequency)
ax1.bar(ml_method_types_sorted, numbers_sorted, color='darkblue', edgecolor='black')
ax1.set_ylabel("Number of Papers", fontsize=12)
ax1.tick_params(axis='x', rotation=15)

# Adjust title position to avoid overlap
plt.title("ML Method in Healthcare", fontsize=14, fontweight='bold', pad=30)

# Line chart (cumulative percentage)
ax2 = ax1.twinx()
ax2.plot(ml_method_types_sorted, cumulative_percent, color='red', marker='o', linewidth=2)
ax2.set_ylabel("Cumulative Percentage (%)", fontsize=12)
ax2.yaxis.set_ticks(np.arange(0, 110, 10))

# Add cumulative percentage labels (shifted slightly lower)
for i, val in enumerate(cumulative_percent):
    ax2.text(i, val - 5, f"{val:.1f}%", ha='center', color='red', fontsize=9)

# Add grid and layout adjustment
ax1.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adds extra space on top
plt.show()
```

### Pareto Chart for Supervised ML Methods in Healthcare

```python
# Data (corrected "atural" -> "Natural")
algorithms = [
    "Decision Tree (DT)",
    "K Nearest Neighbors (K-NN)",
    "LASSO Regression",
    "Logistic Regression (LR)",
    "Multiple Linear Regression (MLR)",
    "Naive Bayes Classifier (NBC)",
    "Natural Language Processing (NLP)",
    "Random Forest (RF)",
    "Support Vector Machine (SVM)",
]
counts = np.array([8, 8, 1, 6, 5, 8, 2, 5, 5])
```

```python
# Sort in descending order for Pareto chart
order = np.argsort(counts)[::-1]
alg_sorted = np.array(algorithms)[order]
counts_sorted = counts[order]

# Calculate cumulative percentage
cum = np.cumsum(counts_sorted)
cum_pct = 100 * cum / cum[-1]
```

```python
# Create Pareto chart
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar chart (frequencies)
bars = ax1.bar(range(len(alg_sorted)), counts_sorted, color='darkblue', edgecolor='black')
ax1.set_ylabel("Number of Papers", fontsize=12)

# Place algorithm names below the x-axis, centered to each bar
ax1.set_xticks(range(len(alg_sorted)))
ax1.set_xticklabels(alg_sorted, rotation=25, ha='right', fontsize=9)

# Title with spacing
plt.title("Supervised ML Algorithms in Healthcare", fontsize=14, fontweight='bold', pad=30)

# Cumulative percentage line (right axis)
ax2 = ax1.twinx()
ax2.plot(range(len(alg_sorted)), cum_pct, color='red', marker='o', linewidth=2)
ax2.set_ylabel("Cumulative Percentage (%)", fontsize=12)
ax2.set_ylim(0, 105)
ax2.yaxis.set_ticks(np.arange(0, 110, 10))

# Annotate cumulative % labels BELOW each point
for i, v in enumerate(cum_pct):
    ax2.text(i, v - 7, f"{v:.1f}%", ha='center', va='top', fontsize=9, color='red')

# Grid + layout adjustment
ax1.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout(rect=[0, 0, 1, 0.93])
plt.show()
```
