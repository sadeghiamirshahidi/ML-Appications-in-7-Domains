```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    'Supervised': {
        'DT': 8,
        'K-NN': 8,
        'LAR': 1,
        'LR': 6,
        'ML)': 5,
        'NBC': 8,
        'NLP': 2,
        'RF': 5,
        'SVM': 5
    },
    'Unsupervised': {
        'GMM': 1,
        'HC': 1,
        'KMC': 2,
        'SOM': 1,
        'SVR': 1
    },
    'Deep Learning': {
        'ANN': 3,
        'CNNs': 6,
        'RNNs': 1
    }
}

# Convert to DataFrame (rows = algorithms, columns = method types)
df = pd.DataFrame(data).fillna(0)

# Matplotlib default colors (matching your figure)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# Plot stacked bar chart
ax = df.T.plot(
    kind='bar',
    stacked=True,
    figsize=(10, 6),
    colormap=None
)

# Manually apply colors to stacks
for i, bar in enumerate(ax.containers):
    for patch in bar:
        patch.set_facecolor(colors[i % len(colors)])

# Labels and formatting
plt.title('Stacked Bar Chart of ML Algorithms by Method Type', fontsize=14)
plt.xlabel('ML Method Type')
plt.ylabel('Number of Studies')
plt.xticks(rotation=0)
plt.legend(title='Algorithms', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
